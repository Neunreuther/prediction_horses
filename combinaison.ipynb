```python
# Initialisation du compteur
compteur = 0

# Exemple de boucles imbriquées pour générer des combinaisons
for i in range(1, 10):     # Disons que tu veux des valeurs entre 1 et 3 pour cet exemple
    for j in range(i+1, 10):
      for k in range(j+1, 10):

        # Incrémentation du compteur
        compteur += 1
        # Affichage de la combinaison avec le numéro
        print(f"Combinaison {compteur}:   (  {i}  {j}  {k}   0)")

# Affichage du nombre total de combinaisons
print(f"Nombre total de combinaisons: {compteur}")
```
